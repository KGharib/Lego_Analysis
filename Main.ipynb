```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
# Reading all csv files and assinging them.
inv = pd.read_csv('inventories.csv')
inv_parts = pd.read_csv('inventory_parts.csv')
inv_sets = pd.read_csv('inventory_sets.csv')
parts = pd.read_csv('parts.csv')
parts_cat = pd.read_csv('part_categories.csv')
sets = pd.read_csv('sets.csv')
themes = pd.read_csv('themes.csv')
colors = pd.read_csv('colors.csv')

dataframes = [inv, inv_parts, inv_sets, parts, parts_cat, sets, themes, colors]

def print_df(dataframes):
    for df in dataframes:
        print(display(df.head()))
        df.info()
        print()
        print('------------------------------------------------------------------------------------')
print_df(dataframes)
```

## Changing Variable Names
Changing variable names in our different dataframe to match schema realtionships (https://www.kaggle.com/rtatman/lego-database?select=downloads_schema.png)

```python
def Change_column_name(df, old, new):
    df.rename(columns={old:new}, inplace=True)

#changing inventory_id --> id    
Change_column_name(inv_sets, 'inventory_id', 'id')
Change_column_name(inv_parts, 'inventory_id', 'id')
#changing id in colors --> color_id
Change_column_name(colors, 'id', 'color_id')
#changing name in color to color_name(as many columns with name)
Change_column_name(colors, 'name', 'color_name')
#changing id in themes --> theme_id
Change_column_name(themes, 'id', 'theme_id')
#changing id in part_cat --> part_cat_id
Change_column_name(parts_cat, 'id', 'part_cata_id')
#changing name in sets to sets_name(as many columns with name)
Change_column_name(sets, 'name', 'sets_name')
#changing name in themes to themes_name(as many columns with name)
Change_column_name(sets, 'name', 'themes_name')
#changing name in parts_cat to parts_cat_name(as many columns with name)
Change_column_name(sets, 'name', 'parts_cat_name')

```

```python
color_inv = pd.merge(inv_parts,colors, on='color_id')
color_inv = color_inv[(color_inv['color_name'] != '[No Color]') & (color_inv['color_name'] != 'Unknown')]
print(color_inv['color_name'].unique())
color_inv.head()
```

```python
cf = color_inv.groupby(['color_name','rgb']).agg(count=('color_name', 'count')).reset_index()
cf.head()
```

```python
cf = cf[cf['count'] > 5000]
cf = cf.sort_values('count', ascending=False)
cf
```

```python
fig, ax = plt.subplots(figsize=(30, 15))
# ax.set_facecolor('beige')
plt.bar(cf['color_name'], cf['count'], color='#'+cf['rgb'], edgecolor='black')
```
